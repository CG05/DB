# Block Quote

- 3단계까지 구성 된다.

- 안녕
    * 반갑
        + 잘가

        - 잘가

- 안녕

        - 잘가

                - 굿모닝

# 강조
```
- '*', '_' 하나를 사용하면 기울임꼴
- '**', '__' 을 사용하면 굵은 글꼴
- '~~' 을 사용하면 취소선
- <u> 를 사용하면 밑줄 표시
- <sup> 윗첨자
- <sub> 아랫첨자
```

*안녕*   
_안녕_   
**안녕**   
__안녕__   
~~안녕~~   
<u>안녕</u>   
O(*n*<sup>2</sup>)   
O(*n*<sub>2</sub>)

# 표 만들기
- th 구분은 '-' 기호로 구분한다
- column 구분은 '|' 기호로 구분한다.

```
|이름|나이|
|---|---|
|홍길동|22|
```

|이름|나이|
|---|---|
|홍길동|22|

- |:---| : 왼쪽 정렬
- |:---:| : 가운데 정렬
- |---:| : 오른쪽 정렬

```
|left_sort|center_sort|right_sort|
|:-|:-:|-:|
||center||
|left|||
|||right|
```
