# 5\. reduce 함수
- 함수를 사용해서 iterable을 집계연산을 통해 하나의 값을 반환한다.
- initializer 는 주어지면 첫번째 인자로 추가된다.
```
구문
functools.reduce(function, iterable[, initializer])
```

```
[ 1 , 2 , 3 , 4 , 5]
add(1,2)
    add)3,3)
       add)6, 4)
          add(10, 5)
```        

````python
# 합 구하기
data = [1,2,3,4]

def total(numbers):
    result = 0
    for number in numbers:
        result += number # 값 누적
        print(f'result : {result}') # 중간 출력
    return result

total(data)
````

````python
# reduce 함수 사용
from functools import reduce

def add(x , y):
    return x + y

reduce(add, [1,2,3,4])
````

````python
def func(a, b):
    print((a, b))
    a.append(b ** 2)
    return a

reduce(func, [1,2,3], [])
````
