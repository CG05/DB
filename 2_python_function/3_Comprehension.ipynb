# 1\. Comprehension 종류
- list Comprehension
- dict Comprehension
- set Comprehension
- 각 list, dict, set for문 사용이 가능하다.

## List Comprehension
```
구문
[ 표현식 for 항목 in 반복객체 (if 조건) ]
```

````python
[i for i in range(10)]
````

````python
dataset = [1,2,3,4]
[num*3 for num in dataset]
````

````python
[num*3 for num in dataset if num%2 == 0]
````

````python
[num * 3 if num % 2 == 0 else num for num in dataset]
````

- for 문 2개 이상 사용도 가능하다.

````python
[
    f'{i} - {j}'
    for i in range(2, 10, 2)
    for j in range(1, 10, 2)
]
````

# 문제
2 ~ 9단까지 list 에 저장해서 출력 해보세요.
```
['2 x 1 = 2',
 '2 x 2 = 4',
 ...
 '9 x 7 = 63',
 '9 x 8 = 72',
 '9 x 9 = 81']
 ```

````python
[
    f'{i} x {j} = {i*j}'
    for i in range(2,10)
    for j in range(1,10)
]
````

````python
# 0 으로 초기화 된 list 
[0 for i in range(10)]
````

````python
dataset = [1, -2, 3, -4, 5]

[
    x ** 2
    for x in dataset
    if x ** 2 > 10
]
````

````python
[
    y for y in [x ** 2 for x in dataset] if y > 10
]
````

## Set Comprehension
- list Comprehension 과 거의 동일한 구문을 사용한다.
```
구문
[ 표현식 for 항목 in 반복객체 (if 조건) ]
```

````python
result = [num % 3 for num in range(10) if num % 3 == 2]
result
````

````python
# set의 특징은 중복된 값은 하나만 저장한다.
result = {num % 3 for num in range(10) if num % 3 == 2}
result
````

## Dict Comprehension
```
구문
[ 키:값 for 항목 in 반복객체 (if 조건) ]
```

````python
{
    n:n**2 for n in range(5)
}
````

````python
{
    x:list(range(x))
    for x in range(1, 6)
}
````
