# sort()

- sort() 함수는 숫자 뿐 아니라 문자들도 정렬을 해 줄 수 있다.

````python
animals = ['dog', 'shark', 'cat', 'bird']
animals.sort()
````

````python
animals
````

````python
animals.append("코끼리")
animals.append("하마")
animals.append("사자")
animals
````

````python
# 숫자 > 영문 > 한글
animals.sort()
animals
````

````python
animals.sort(reverse=True)
animals
````

````python
animals[1:]
````

````python
animals[:4]
````

````python
animals[:]
````

````python
# 슬라이싱 [시작:끝-1:스텝]
animals[::2]
````

````python
animals[1:5:3]
````

````python
data = [x for x in range(10)]
data
````

````python
data[2::3]
````

````python
data[::1]
````

````python
data[-1]
````

````python
data[::-1]
````

````python
data[::-2]
````

# sorted() 함수
```
구문
sorted(iterable, key=key, reverse=reverse)
```
- iterable : sort 대상 (list, dict, tuple 등... )
- key : 대소비교를 하는 함수(기본값은 None)
- reverse : False 오름차순(기본값), True 내림차순

````python
a = ['b', 'g', 'a', 'd', 'c', 'f', 'h', 'e']
a
````

````python
sorted(a)
````

````python
a
````

````python
sorted(a, reverse=True)
````

## dict 정렬 및 출력
- 기본적으로 dict 는 순서가 없다.
- key 로 정렬하고 출력하는 방법
- value 로 정렬하고 출력하는 방법

````python
dict = {'b':5, 'c':3, 'a':2}
dict
````

````python
# dict.items() : 키와 값을 반환
# 키로 정렬
dict_items = dict.items()
sorted(dict_items)
````

````python
# 값으로 정렬 - 내림차순
sorted(dict.items(), key=lambda x:x[1], reverse=True)
````

````python
sorted(dict.items(), key=lambda x:x[0], reverse=True)
````

````python
# 키 오름차순
sorted(dict)
````

````python
# 값 오름차순
sorted(dict, key=dict.get)
````

````python
[
    (key, dict[key])
    for key in sorted(dict, key=dict.get)
]
````
