# NumPy 시각화
- matplotlib 모듈 이용하여 그래프 표현을 할 수 있다.
- 데이터 시각화는 데이터 분석, 머신러닝에서 매주 중요한 프로세스다.
- 시각화를 통해 데이터 추세나 분포 등에 대한 인사이트와 전략을 세울수 있도록 해야 한다

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 11)
x
```

```python
y = x ** 2 + x + 2
y
```

## 그래프 출력하기
 - plot함수 (선 그래프), scatter(점 그래프), hist(히스토그램) 등 사용
   - 함수의 parameter 혹은 plt의 다른 함수로 그래프 형태 및 설정을 변경 가능
   - 기본적으로, x, y에 해당하는 값이 필요

```python
plt.plot(x, y) # line plot
```

```python
plt.scatter(x, y) # scatter plot - 점으로 표현된 그래프
```

```python
y = x ** 2 + x + 2 + np.random.randn(11)
y
```

```python
plt.plot(x, y)
```

```python
plt.scatter(x, y)
```

## 그래프에 레이블, 타이틀 추가
- xlabel(), ylabel(), title()

```python
plt.xlabel('X values') # x 축의 이름
plt.ylabel('Y values') # y 축의 이름
plt.title('X-Y relation') # 그래프의 이름
plt.plot(x, y)
```

## grid 추가

```python
plt.xlabel('X values') # x 축의 이름
plt.ylabel('Y values') # y 축의 이름
plt.title('X-Y relation') # 그래프의 이름
plt.grid(True) # 그래프 좌표 표시
plt.plot(x, y)
```

## x, y 축 표시범위 지정
- xlim(), ylim()

```python
plt.xlabel('X values') # x 축의 이름
plt.ylabel('Y values') # y 축의 이름
plt.title('X-Y relation') # 그래프의 이름
plt.grid(True) # 그래프 좌표 표시
plt.xlim(0, 20) # x 의 표시 범위(시작, 끝)
plt.ylim(0, 200) # y 의 표시 범위(시작, 끝)
plt.plot(x, y)
```

# plot 함수 parameters

## keyword parameter 이용하여 모든 속성 설정
 - https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot
 - color

 - markerfacecolor
 - markersize 등등

```python
plt.plot(x, y, color='green', linestyle='--', marker='p', markerfacecolor='red', markersize=15, linewidth=5)
```

## subplot으로 여러 그래프 출력하기
 - subplot함수로 구획을 구별하여 각각의 subplot에 그래프 출력

```python
plt.subplot(2, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(2, 2, 2)
plt.plot(x, x** 3, 'g')

plt.subplot(2, 2, 3)
plt.plot(y, x, 'm')

plt.subplot(2, 2, 4)
plt.plot(x, np.exp(x), 'b')
```

- 여러개의 그래프 (ptot) 을 하나에 담을수 있다!

- subplot 은 여러개의 data feature 와 target 사이의 관계를 한번에 대조하면서 보고자 할때 많이 사용

## hist함수 사용
 - histogram 생성 : 데이터 변량의 분포를 표현하는 그래프
 - bins로 historgram bar 개수 설정

```python
data = np.random.randint(1, 100, size=200)
data
```

```python
plt.hist(data, bins=10)
plt.xlabel('val')
plt.ylabel('cnt')
plt.grid(True)
plt.show()
```

```python
plt.hist(data, bins=20)
plt.xlabel('val')
plt.ylabel('cnt')
plt.grid(True)
plt.show()
```

```python
plt.hist(data, bins=20, alpha=0.4)
plt.xlabel('val')
plt.ylabel('cnt')
plt.grid(True)
plt.show()
```
