### 도전] 남성과 여성의 생존률은 ?

> [결과 예시]<br>
> 남성생존률 0.18890814558058924<br>
> 여성생존률 0.7420382165605095

```python
df[df['Sex'] == 'male']
```

```python
male_mask = df['Sex'] == 'male'
female_mask = df['Sex'] == 'female'
```

```python
df[male_mask]['Survived'] == 1
```

```python
df[male_mask]['Survived'].mean()
```

```python
df[female_mask]['Survived'].mean()
```

```python
len(df[(df['Sex']=='male') & (df['Survived'] == 1)]) / len(df[df['Sex'] == 'male'])
```
