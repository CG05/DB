# Row, Column 추가 / 삭제

```python
import pandas as pd
import numpy as np 

df = pd.read_csv('data/titanic.csv')
df.head()
```

## 새 column 추가하기
 - [] 사용하여 추가하기
 - insert 함수 사용하여 원하는 위치에 추가하기

```python
df['Name']
```

```python
df['Age']
```

```python
df['Age_double']
```

```python
# [] 로 column 을 추가할 때는 없는 column 명을 지정
df['Age'] * 2
```

```python
df['Age_double'] = df['Age'] * 2
```

```python
df[['Age', 'Age_double']]
```

```python
df
```

```python
df['Age_triple'] = df['Age_double'] + df['Age']
df.head()
```

- 위와 같이 기존에 존재하는 컬럼을 새로운 컬럼으로 생성할 수 있다.
- 새로 추가되는 컬럼은 반드시 맨뒤에 삽입 된다.

### insert() 
원하는 위치에 컬럼 추가

```python
# insert(위치, 'column명', 데이터)
df.insert(10, 'Fare10', df['Fare'] / 10)
df
```

## column 삭제하기
 - drop 함수 사용하여 삭제
   - axis= 사용
   - 리스트를 사용하여 멀티플 삭제 가능 

```python
# DataFrame 은 2차원 이기 때문에 axis = 0 행, axis = 1 열 을 삭제한다.
# drop('컬럼명', axis = 1)
df.drop('Age_triple', axis = 1)
```

```python
df.drop('Age_triple', axis=1, inplace=True)
```

```python
df.head()
```

```python
df.drop(['Age_double', 'Fare10'], axis=1, inplace=True)
```

```python
df.head()
```

#### 행 삽입

```python
# DataFrame.insert() : 컬럼을 위한 함수
# Numpy.insert() : 함수를 이용하면 row 값을 넣을 수 있다.

df2.values
```

```python
newData = np.insert(df2.values, 2, values=[0.1,0.2,0.3,0.4], axis=0)
newData
```

```python
newIndex = np.insert(df2.index, 2, '하')
newIndex
```

```python
pd.DataFrame(
    data = newData,
    columns = df2.columns,
    index = newIndex
)
```

```python
# 방법 : concat 과 iloc 을 사용하여 삽입
```

```python
df2.iloc[:2]
```

```python
df2.iloc[2:]
```

```python
new_df = pd.DataFrame({'a':0.01, 'b':0.02, 'c':0.03, 'd':0.04}, index=['하'])
new_df
```

```python
pd.concat([
    df2.iloc[:2],
    new_df,
    df2.iloc[2:]
])
```

```python
# 방법 : transpose 이용한 삽입
df2_t = df2.T
df2_t
```

```python
df2_t.insert(2, '하', [0.1, 0.2, 0.3, 0.4])
df2_t.T
```
