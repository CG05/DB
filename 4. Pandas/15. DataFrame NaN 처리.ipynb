# NaN 처리 

NaN(Not A Number)
- 결측치(missing value)라고 한다.
- 결측치가 발생하는 이유는 데이터의 누락, 또는 소실 등이 원인이 된다.
- 즉 값이 없다는 뜻이다. (숫자가 아니라는 의미가 아니다.)
- NaN은 반드시 데이터 분석전에 데이터를 확인하여 NaN을 정리 해야 한다.
- 이것을 우리는 데이터 전처리 라고 한다.

NaN 값을 처리하는 방법
- 결측치를 제거 하는 방법
- 다른값으로 대체 하는 방법

```python
import pandas as pd
import numpy as np

df = pd.read_csv('data/titanic.csv')
df.head()
```

## NaN 값 확인
 - describe, info함수를 통하여 개수 확인
 - isna함수를 통해 boolean 타입으로 확인

```python
# 숫자데이터만 나옴
df.describe()
```

```python
df.info()
```

```python
# DataFrame.isna() : NaN 값이 있으면 True, 없으면 False 나타내준다.
df.isna()
```

```python
df['Age'].isna()
```

```python
df[df['Age'].isna()]
```

```python
# 컬럼별 결측치 수
df.isna().sum()
```

```python
# 전체 결측치 수
df.isna().sum().sum()
```

## NaN 처리 방법
 - 데이터에서 삭제
   - dropna 함수 
 - 다른 값으로 치환
   - fillna 함수

#### NaN 데이터 삭제

```python
df.shape
```

```python
df.dropna().shape
```

```python
df.dropna()
```

```python
# dropna(axis=0) : 기본적으로 NaN 이 포함된 row 을 삭제 한다.
# NaN이 하나라도 있는 행은 삭제된다는 의미다.
```

```python
# dropna(axis=1) : 컬럼에 NaN이 있으면 컬럼 삭제한다.
# NaN이 하나라도 있는 컬럼은 삭제된다는 의미다.
df.dropna(axis=1)
```

```python
# 특정 컬럼의 NaN 값이 있는 행만 삭제
df.dropna(subset=['Age'])
```

```python
df.dropna(subset=['Age', 'Cabin'])
```

### NaN 값 대체하기
 - fillna()
 - 평균으로 대체하기
 - 생존자/사망자 별 평균으로 대체하기

```python
# 대표적인 수치 데이터의 대체값은 평균값으로 한다.
df['Age'].mean()
```

```python
df[df['Age'] > 0]
```

```python
# 인덱스명
df.loc[888]
```

```python
# 위치값
df.iloc[888]
```

```python
# NaN가 제외한 평균값
df['Age'].mean()
```

```python
len(df['Age']) - df['Age'].isna().sum()
```

```python
df['Age'].fillna(df['Age'].mean())
```

## 도전] Age NaN 값을 생존자 / 사망자 별 평균으로 대체하기

```python
life_mean = df[df['Survived'] == 1]['Age'].mean()
life_mean
```

```python
death_mean = df[df['Survived'] == 0]['Age'].mean()
death_mean
```

```python
df['Age'].fillna(life_mean)
```

```python
df['Age'].fillna(death_mean)
```

```python
df[df['Survived'] == 1]['Age'].fillna(life_mean)
```

```python
df[df['Survived'] == 0]['Age'].fillna(death_mean)
```

```python
df[df['Survived'] == 1, 'Age'] = df[df['Survived'] == 1]['Age'].fillna(life_mean)
```

```python
df.loc[df['Survived'] == 1, 'Age'] = df[df['Survived'] == 1]['Age'].fillna(life_mean)
```

```python
df['Age'].isna().sum()
```

```python
df.loc[df['Survived'] == 0, 'Age'] = df[df['Survived'] == 0]['Age'].fillna(death_mean)
```

```python
df['Age'].isna().sum()
```
