## 숫자형(Numerical Type) 데이터
 - 연속성을 띄는 숫자로 이루어진 데이터
   - 예) Age, Fare 등
 - 대소비교, 산술연산 (평균, 합계, 중간값) 의미있슴
 
## 범주형(Categorical Type) 데이터
 - 연속적이지 않은 값(대부분의 경우 숫자를 제외한 나머지 값)을 갖는 데이터를 의미
   - 예) Name, Sex, Ticket, Cabin, Embarked 
 - 어떤 경우, 데이터타입(dtype)이 숫자형 타입이라 할지라도 개념적으로 범주형으로 처리해야할 경우가 있음
   - 예) Pclass, Survived   
 - 대소비교, 산술연산(평균, 합계, 중간값...) 의미없슴

```python
# 비교 연산자를 이용하여 데이터를 처리 할 수 있다.
df[df['Age'] >= 30]
```

```python
df[df['Age'] == 30]
```

```python
df[df['Sex'] == 'male']
```

```python
df[df['Sex'] > 'male']
```

```python
df['Pclass'].mean()
```

```python
# 숫자형 데이터 --> 범주형 데이터 변환이 가능
# 범주형 데이터 --> 숫자형 데이터 변환이 가능 (One-hot encoding)
```

### apply()  함수 사용!
변환로직을 '함수' 로 만든뒤 적용

```python
def age_categorical(age):
    return int(age/10) * 10
```

```python
age_categorical(22)
```

```python
age_categorical(43)
```

```python
import math

def age_categorical(age):
    if math.isnan(age):
        return -1
    return int(age/10) * 10
```

```python
age_span = df['Age'].apply(age_categorical)
age_span
```

```python
# 출력 순
age_span.unique()
```

```python
# 많은 순
age_span.value_counts()
```

```python
df['Age_span'] = age_span
df
```
