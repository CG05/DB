# One-hot encoding
범주형(Categorical) 데이터 전처리하기

- 범주형 데이터는 연산이 불가능하다. 또는 의미가 없다.
- One-hot encoding 을 사용하는데 데이터의 값이 컬럼이 되고, 값이 있으면 1, 없으면 0 으로 처리

```
color = ['red','blue','black', 'red', 'black']
```
--------------------------------
One_Hot encoding 을 처리 하면 
```
red = [1, 0, 0, 1, 0]
blue = [0, 1, 0, 0, 0]
black = [0, 0, 1, 0, 1]
```

## One-hot encoding
 - 범주형 데이터는 분석단계에서 계산이 어렵기 때문에 숫자형으로 변경이 필요함
 - 범주형 데이터의 각 범주(category)를 column레벨로 변경
 - 해당 범주에 해당하면 1, 아니면 0으로 채우는 인코딩 기법
 - pandas.get_dummies 함수 사용
   - drop_first : 첫번째 카테고리 값은 사용하지 않음

````python
# Pandas.get_dummies() : 문자형 데이터를 One-hot encoding 을 처리한다.
# Name, Cabin, Embarked 를 처리했다.
# 문제는 Name 같은 경우 이름마다 컬럼을 만들었기 때문에 많은 컬럼들이 생긴다.
pd.get_dummies(df)
````

````python
pd.get_dummies(df, columns=['Pclass', 'Sex', 'Embarked'])
````

````python
# Pandas.get_dummies(데이터, columns=[], drop_first=True(False))
# drop_first : 첫번째 column 이 필요 없을 때 사용 한다.

pd.get_dummies(df, columns=['Pclass', 'Sex', 'Embarked'], drop_first=True)
````

````python
# Sex 같은 경우 male 이랑 female 밖에 없다.
# male 이 아니면 female 이기 때문에 하나의 컬럼만 만들어도 되기 때문이다.
````

````python
# 머신 러닝을 하다보면 범주형 데이터를 수치화했을 때 1, 0 으로 나올때가 대부분이다.
# 하지만 가끔 0.98, 0.01 이런식으로 나오는 경우도 있다.
# 그럴때는 1에 가까운 것이 True 값이다.
````

### 리스트의 형태로 one-hot encoding 만들기
- 컬럼을 추가하는 방식이 아닌 list 형태의 one-hot encoding
- 수동으로 만들어야 하는 경우도 제법 있다

````python
df['Pclass']
````

````python
one_hot = { 1:[1,0,0], 2:[0,1,0], 3:[0,0,1] }
````

````python
df['Pclass'].apply(lambda x : one_hot[x])
````

````python
# 하지만 나중에 딥러닝 keras 에서는 to_categorical 등을 통해서 간단히 범주형 데이터를 변경할 수 있다.
````

# 실습 문제

````python
import pandas as pd
data = { 'country': [ 'Belgium', 'France', 'Germany', 'Netherlands', 'United Kingdom'],
		'population' : [11.3,64.3, 81.3, 16.9, 64.9],
		'area' : [30510, 671308, 357050, 41526, 244820],
		'capital' : ['Brusseis', 'Paris', 'Berlin', 'Amsterdam' , 'London'] }
	
countries = pd.DataFrame(data)
countries = countries.set_index('country')
countries	
````

문제 1. 인구 밀도를 의미하는 'density' 칼럼을 추가하세요 (주의 : 현재 'population' 칼럼은 10만 단위로 표기 되어 있습니다.)
```
countries['density'] = countries[?] * 100000 / countries[?]
countries
```

````python
countries['density'] = countries['population'] * 100000 / countries['area']
countries
````

문제 2. 인구 밀도가 30을 초과하는 국가(Country)의 수도(capital)와 인구(population)을 선택해 주세요

````python
countries.loc[countries['density'] > 30, ['capital', 'population']]
````

````python
density30 = countries[countries['density'] > 30]
density30[['capital', 'population']]
````

````python
density_mask = countries['density'] > 30
countries[density_mask][['capital', 'population']]
````

문제 3. 'density_ratio' 칼럼을 추가해주세요. (density_ratio = 인구밀도 / 평균 인구 밀도)

````python
countries['density'].mean()
````

````python
countries['density_ratio'] = countries['density'] / countries['density'].mean()
countries
````
