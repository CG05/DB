# 남미대륙 (South America) 소속국가들중 평균면적보다 작은 나라들은?
- '이름' 과 '면적' 출력
- 면적이 작은것부터 오름차순으로 정렬
- index 는 1부터 시작
- ./out/country_3.csv 파일로 저장

[결과예시]


```python
sa_mask = df['Continent'] == 'South America'
sa_mask
```

```python
df[sa_mask]
```

```python
# 평균값
sa_mean_surface = df[sa_mask]['SurfaceArea'].mean()
```

```python
df[sa_mask]['SurfaceArea'] < sa_mean_surface
```

```python
df[sa_mask][df['SurfaceArea'] < sa_mean_surface]
```

```python
# sa_mask = South Ameria 일 경우 True 아니면 False
df_result2 = df[(sa_mask) & (df['SurfaceArea'] < sa_mean_surface)][['Name', 'SurfaceArea']]
df_result2
```

```python
df_result2.sort_values('SurfaceArea', inplace=True)
```

```python
df_result2.reset_index(drop=True, inplace=True)
```

```python
df_result2.index += 1
df_result2
```

```python
df_result2.to_csv('data/out/country_3.csv')
```
