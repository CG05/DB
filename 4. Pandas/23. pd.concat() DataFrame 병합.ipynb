# pd.concat()
DataFrame 병합시키기

```python
import pandas as pd
import numpy as np
```

## pd.concat() 함수 사용하여 dataframe 병합하기
 - pandas.concat 함수
 - 축(axis)을 따라 dataframe을 병합 가능
   - axis = 0 → 행단위 병합 (디폴트)
   - axis = 1 → 열단위 병합

```python
df1 = pd.DataFrame({'key1' : np.arange(10), 'value1': np.random.randn(10)})
df2 = pd.DataFrame({'key1' : np.arange(10), 'value1': np.random.randn(10)})
```

```python
df1
```

```python
df2
```

### column / index 가 같은 경우  pd.concat 하기()

```python
# Pandas.concat([데이터프레임 list]) : 행단위로 DataFrame 연결
pd.concat([df1, df2])
```

```python
# ignore_index=True : 기존 인덱스를 무시하고 새로운 인덱스로 적용
pd.concat([df1,df2], ignore_index=True)
```

```python
# axis=1 : 열로 DataFrame을 연결한다.
pd.concat([df1, df2], axis=1)
```

### column / index 가 다른 경우  pd.concat 하기()

```python
# 컬럼명이 다른 경우
df3 = pd.DataFrame({ 'key2': np.arange(10), 'value2':np.random.randn(10)})
df3
```

```python
# 없는 컬럼은 NaN으로 채워준다.
pd.concat([df1, df3])
```

```python
df2.index += 10
df2
```

```python
# 없는 index의 값과 컬럼은 NaN 값으로 채워진다.
pd.concat([df1, df2], axis=1)
```

```python
df3.index += 10
df3
```

```python
pd.concat([df1, df3])
```

```python
pd.concat([df1, df3], axis=1)
```
