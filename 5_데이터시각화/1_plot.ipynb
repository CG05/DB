# Matplotlib
https://matplotlib.org/

#### matplotlib 시각화 예제 갤러리
https://matplotlib.org/stable/gallery/index.html

- 파이썬의 자료(DataFrame, Series 등..)을  차트(chart) 나 플롯(plot) 으로 시각화(visualization) 하는 모듈이다.
- Matplotlib는 다음과 같은 정형화된 차트나 플롯 이외에도 저수준 api를 사용한 다양한 시각화 기능을 제공한다.
- Matplotlib 시각화 함수
  - 라인 플롯(line plot)
  - 스캐터 플롯(scatter plot)
  - 컨투어 플롯(contour plot)
  - 서피스 플롯(surface plot)
  - 바 차트(bar chart)
  - 히스토그램(histogram)
  - 박스 플롯(box plot)

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

# Plot

## Line plot
가장 간단한 플롯,  선을 그리는 라인 플롯<br>
데이터가 시간, 순서 등에 따라 어떻게 변화하는지 보여주기 위한 용도

http://Matplotlib.org/api/pyplot_api.html#Matplotlib.pyplot.plot

## array -> line plot

```python
arr = np.array([10,30,14,40])
arr
```

```python
# list 에서 인덱스 번호는 x축, 값은 y축
plt.plot(arr)
```

```python
arr = np.array([
    [10,30,14,20],
    [20,50,40,10],
    [30,20,10,5]
])
arr
```

```python
# 이차원 배열에서는 각 행의 인덱스 번호가 x축, 값은 y축
plt.plot(arr)
plt.show()
```

```python
# 3차원 ndarray는 불가능
x = np.arange(0,1,0.01)
y = x ** 2
plt.plot(x,y)
plt.show()
```

## Series -> line plot

```python
s = pd.Series(np.random.randn(10).cumsum(), index=np.arange(0,100,10))
s
```

```python
# Series 는 1차원이다.
# 인덱스는 x축, 값은 y축
plt.plot(s)
plt.show()
```

```python
# pandas 객체는 plot() 함수 제공
s.plot()
```

## DataFrame -> line plot

```python
# DataFrame 은 2차원
df = pd.DataFrame(np.random.randn(10, 4).cumsum(axis=0),
                  columns=['A','B','C','D'],
                  index=np.arange(0,100,10))
df
```

```python
# index 는 x축, 값은 y축
plt.plot(df)
# 라인구분 label
plt.legend(df.columns)
plt.show()
```

```python
df.plot()
```

```python
df['B'].plot()
```

```python
plt.plot(df['B'])
plt.show()
```

```python
df[['B','A']].plot()
```

```python
plt.plot(df[['B','A']])
plt.show()
```

## Histogram 그리기
도수분포표의 하나, 가로축이 계급, 세로축에 도수,

```python
s3 = pd.Series(np.random.normal(0, 1, size=200))
s3
```

```python
# x 축은 값, y 축 분포(수)
s3.hist()
```

```python
plt.hist(s3)
plt.show()
```

```python
# 바를 bin 이라고 한다.
s3.hist(bins=50)
```

```python
plt.hist(s3, bins=50)
plt.grid()
plt.show()
```

## Scatter plot 그리기
산점도 : 산점도의 경우에는 서로 다른 두 개의 독립변수에 대해 두 변수가 어떤 관계가 있는지 살펴보기 위해 사용된다.


```python
x1 = np.random.normal(1, 1, size=(100, 1))
x1
```

```python
x2 = np.random.normal(-2, 4, size=(100,1))
x2
```

```python
X = np.concatenate((x1, x2), axis=1)
X
```

```python
df3 = pd.DataFrame(X, columns=['x1', 'x2'])
df3
```

```python
plt.scatter(df3['x1'], df3['x2'])
plt.show()
```

## 여러 선 그리기

```python
x = np.arange(0, 1, 0.01)

plt.plot(
    x, x, 'r--', 
    x, x ** 2, 'k--',
    x, x ** 3, 'b--'
)
plt.show()
```
