### plot 의 눈금(tick), 레이블(label), 범위

```python
fig = plt.figure()
```

```python
ax = fig.add_subplot(1,1,1)
```

```python
ax.plot(np.random.randn(1000).cumsum())
fig
```

```python
fig, ax = plt.subplots(1,1)
ax.plot(np.random.randn(1000).cumsum())
```

```python
# set_xticks : x label이 나타나는 값을 지정할 수 있다.
# set_yticks : y label이 나타나는 값을 지정할 수 있다.
np.random.seed(1000)
fig, ax = plt.subplots(1,1)
ax.plot(np.random.randn(1000).cumsum())
# ax.set_xticks([0, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000])
# ax.set_xticks([0, 250, 500, 750, 1000])
ax.set_xticks(np.arange(0, 1000, 300))
ax.set_yticks([-30, -10, 10])
plt.show()
```

```python
# 기존 xticks 는 출력할 값과 위치를 지정한다.
# set_xticklabels() : 출력할 값을 꾸민다.
fig, ax = plt.subplots(1,1, figsize=(8,8))
ax.plot(np.random.randn(1000).cumsum())
ax.set_xticks([0, 250, 500, 750, 1000])

# ratation : 양수면 반시계방향, 음수면 시계방향 label 을 기울인다.
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'],
                            rotation = 30, fontsize='large')

# 그래프에 제목
ax.set_title('random walk plot')

# label
ax.set_xlabel('Stages')
ax.set_ylabel('Values')

# 눈금 - grid(True) : 눈금이 생김, grid(False) : 눈금이 없어진다.
ax.grid()
plt.show()
```

```python
np.random.seed(50)
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum(), 'k', label='one')
ax.plot(np.random.randn(1000).cumsum(), 'b--', label='two')
ax.plot(np.random.randn(1000).cumsum(), 'r.', label='three')

# legend : 범례 - 각 선의 이름 출력
# legend(loc='best') : loc='best' 는 없어도 범례를 보기 제일 좋은 곳에 위치시킨다.
ax.legend()

# x축, y축 범위 지정
ax.set_xlim([0, 1000])
ax.set_ylim([-100, 100])

plt.show()
```
