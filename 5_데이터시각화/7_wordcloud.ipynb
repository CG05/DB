wordcloud 라이브러리는 matplotlib 과 함께 사용됨

```python
import wordcloud
```

```python
import matplotlib.pyplot as plt
```

```python
words = {
    'Python':10,
    'Java':5,
    'C':7,
    'C++':9,
    'Javascript':8
}
```

```python
wc = wordcloud.WordCloud()
wc
```

```python
cloud = wc.generate_from_frequencies(words)
plt.imshow(cloud)
plt.show()
```

```python
words = {
    '파이썬' : 10,
    '빅데이터':5,
    '웹크롤링':7,
    '인공지능':9,
    '딥러닝':8
}
```

```python
cloud = wc.generate_from_frequencies(words)
plt.imshow(cloud)
plt.show()
```

```python
wc = wordcloud.WordCloud(font_path="c:/Windows/Fonts/malgun.ttf")
cloud = wc.generate_from_frequencies(words)
plt.imshow(cloud)
plt.show()
```

```python
wc = wordcloud.WordCloud(font_path="c:/Windows/Fonts/malgun.ttf",
                         background_color='white', repeat=True)
cloud = wc.generate_from_frequencies(words)
plt.imshow(cloud, interpolation='bilinear')
plt.show()
```
