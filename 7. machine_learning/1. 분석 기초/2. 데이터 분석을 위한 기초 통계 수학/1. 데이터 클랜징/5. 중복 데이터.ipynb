##2. 데이터 중복 확인
> DataFrame에서는 중복 확인을 위해 duplicated 함수를 지원한다. 중복된 데이터가 나타나면 true로 변환시킨다. 

>0번과 1번의 이용거리가 같아 1번 이용거리가 True로 설정되었다. 416~418까지도 중복된 데이터이기때문에 True로 설정되었다.

>두 가지 이상의 옵션을 설정하게 되면 and로 중복을 확인할 수 있다.

>keep 옵션은 기본값이 first이다. first는 첫번째 값을 기준으로 중복된 데이터가 나오면 true로 변환시키는 것이고 last는 중복된 가장 마지막 값을 false로 지정하는 것이다.

keep이 false이면 중복된 데이터를 모두 true로 변환시킨다.

##3. 중복 데이터 제거
>중복된 데이터를 제거하기 위해서는 drop_duplicates() 함수를 이용하면 된다. 사용방법은 중복데이터 확인과 같다.
