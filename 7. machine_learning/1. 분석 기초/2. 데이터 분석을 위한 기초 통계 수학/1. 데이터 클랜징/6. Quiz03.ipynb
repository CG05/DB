###3.3 이상치 함수 만들기

```python

abnormals_s = cty[(cty < lowerfence) | (cty > upperfence)]
abnormals_s
```

>이상치가 여러개 존재하는 경우를 알아보자. 위와 같이 이상치가 존재한다고 가정할 때 다른 값으로 치환하기 위해 replace를 이용할 것이다. replace는 첫 번째 인자의 값을 두 번째 인자의 값으로 변환하는 기능을 한다. 위와 같은 경우 이상치는 999와-50이다 여기서 999는 중복된 데이터임으로 반복을 시킨다면 2번 동작되게 될 것이다. 이를 해결하기 위해 우선 중복 데이터를 제거해 보자

```python
abnormals = abnormals_s.drop_duplicates()
abnormals = abnormals.values
abnormals
```

```python
def getFence(data, gap):
    q1, q2, q3, q4 = data.quantile([0.25, 0.5, 0.75, 1])
    iqr = q3 - q1
    upperfence = q3 + gap*iqr
    lowerfence = q1 - gap*iqr

    return upperfence, lowerfence

def checkAbnormal(data, gap):
    upperfence, lowerfence = getFence(data, gap)
    abnormals_s = data[(data > upperfence) | (data < lowerfence)]

    return abnormals_s

checkAbnormal(cty, 4.5)
```

>변수지정 부분은 추후 가변적으로 변경될 내용을 미리 작성해 둔 것이다. 함수 작성 시 활용하기 위해 변수로 만들어 두었다. 여기서는 굳이 변수 지정을 하지 않아도 상관은 없다.

>이상치 체크 부분은 발생된 이상치의 위치를 얻기 위해 진행했다.

>중복데이터 제거 부분에서는 이상치 데이터 중 중복된 데이터를 제거하기 위한 것이다.

```python
abnormals_s = checkAbnormal(cty, 4.5)
print(abnormals_s)
```

###3.4 이상치 제거 목록 만들기
>3.2 한계치를 이용한 이상치 만들기에서 최고 한계치를 100으로 두어야 된다고 했다. 이를 이용하여 이상치 제거 목록을 만들어 보자

>위에서 만든 목록을 바탕으로 모든 이상치를 치환해 보자 

```python
# iterrows() 함수를 사용하여 데이터프레임의 행을 반복 처리
for idx, row in df.iterrows():
    print('idx:', idx)
    print('row:', row)
```

```python
def getMedian(row, column):
    mask = True
    for c in df.columns:
        if c != column:
            mask = mask & (df[c] == row[c])
    return df[mask][column].median()
    
# 이상치를 median으로 변환하는 함수
def convertAbnormality(column, abnormals):
    for abnormal in abnormals:
        # 이상치 데이터 추출
        target = df[df[column] == abnormal]
        # 이상치 데이터의 행을 반복 처리
        for idx, row in target.iterrows():
            print('idx:', idx)
            print(getMedian(row, column))
            # 이상치를 median으로 변환
            df.loc[idx, column] = getMedian(row, column)
```

>실행결과 다음과 같은 문제가 발생했다.
*        class    cty   hwy
*105  subcompact   25.0  36.0
*106  subcompact  512.0  36.0

>데이트가 2개만 존재하여 (25 + 999)/2로 도심연비가 나오게 되었다. 이를 해결하기 위해 데이터가 2개 이하이면 최소값을 중간값으로 처리하도록 함수를 변경해 적용하자.

```python
# column을 제외한 나머지 column이 같은 row의 median을 구하는 함수
def getMedian(row, column):
    mask = True
    for c in df.columns:
        if c != column:
            mask = mask & (df[c] == row[c])
    # 데이터가 3개 미만인 경우 min값을 반환
    if df[mask][column].size < 3:
        return df[mask][column].min()
    # 데이터가 3개 이상인 경우 median값을 반환
    return df[mask][column].median()
```

```python
def processAbnormality(column, gap):
    # 이상치를 찾을 column의 값만 추출
    targetCol = df[df[column].notnull()][column]
    # 이상치를 찾아서 Series로 반환
    abnormals_s = checkAbnormal(targetCol, gap)
    print("abnormals_s :", abnormals_s)
    # 중복값 제거
    abnormals = abnormals_s.drop_duplicates().values
    # 이상치를 median으로 변환
    convertAbnormality(column, abnormals)
    # 이상치가 제대로 제거되었는지 확인
    print(checkAbnormal(df[df[column].notnull()][column], gap))
```

```python
# NaN 값을 평균 값으로 대체하는 함수
def getMean(row, column):
    mask = True
    for c in df.columns:
        if c != column:
            mask = mask & (df[c] == row[c])
    return df[mask][column].mean()

def fillNan(column):
    # NaN 값이 있는 데이터를 선택합니다.
    nan_datas = df[df[column].isnull()]
    print(nan_datas)
    # 각 NaN 값을 가진 행에 대해 반복합니다.
    for idx, nan_data in nan_datas.iterrows():
        # NaN 값을 평균 값으로 대체합니다.
        df.loc[idx, column] = getMean(nan_data, column)
    # NaN 값이 제대로 대체되었는지 확인합니다.
    print(df[df[column].isnull()])
    # NaN 값이 없다면 제대로 대체된 것입니다.
    if not df[df[column].isnull()].empty:
        # NaN 값이 남아있다면, 남은 NaN 값을 제거합니다.
        df.dropna(subset=[column],inplace=True)
    # NaN 값이 제대로 제거되었는지 확인합니다.
    print(df[df[column].isnull()])
        
fillNan('cty')
```

>위와 같이 간단하게 결측치 변경 함수를 만들었다.
