## 다음 데이터를 이용하여 데이터를 분석하시오
```
df = pd.DataFrame({
    'city': ['부산', '부산', '부산', '부산', '서울', '서울', '서울'],
    'fruits': ['apple', 'orange', 'banana', 'banana', 'apple', 'apple', 'banana'],
    'price': [100, 200, 250, 300, 150, 200, 400],
    'quantity': [1, 2, 3, 4, 5, 6, 7]
})
```

>1. 도시별 가격 평균을 구하시오
>2. 각 도시에서 과일별 가격 평균을 구하시오
>3. 부산에서 가장 비싼 과일은 몇개인가?
>4. 각 도시별로 가장 비싼 가격과 가장 적은 개수를 구하시오

````python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'city': ['부산', '부산', '부산', '부산', '서울', '서울', '서울'],
    'fruits': ['apple', 'orange', 'banana', 'banana', 'apple', 'apple', 'banana'],
    'price': [100, 200, 250, 300, 150, 200, 400],
    'quantity': [1, 2, 3, 4, 5, 6, 7]
})

df
````

````python
# 도시별 가격 평균
df1 = df.groupby('city').agg({'price': 'mean'})
df1.columns = ['price_mean']
df1
````

````python
# 각 도시에서 과일별 가격 평균
df2 = df.groupby(['city', 'fruits']).agg({'price': 'mean'})
df2.columns = ['price_mean']
df2
````

````python
# 부산에서 가장 비싼 과일 개수
df3 = df.groupby('city').agg({'price': 'max', 'quantity': 'count'})
print("부산에서 가장 비싼 과일 개수 : ",df3.loc['부산', 'quantity'])
````

````python
# 각 도시별 가장 비싼 가격과 가장 적은 개수
df4 = df.groupby('city').agg({'price': 'max', 'quantity': 'min'})
df4.columns = ['price_max', 'quantity_min']
df4
````
