## 검색 기간으로 2023년 01월 파일 저장
- 파일의 윗부분은 데이터를 정제해야 한다.

```python
import pandas as pd

df = pd.read_excel('data/2023년 1월 업무추진비 집행내역 공개.xlsx')
df
```

## 불필요한 컬럼 제거
- No, 구분, 비목 제거



```python
df.drop(['구분(시장실만 사용)', '비목', 'No.'], axis=1, inplace=True)
df
```

## 결제 방법에 따른 금액 확인
- 필요한 칼럼만 사용

```python
df.groupby(['결제방법'])['사용금액(원)'].sum()
```

## 일자별 사용 금액 확인

```python
df1 = df.groupby(['사용일'])['사용금액(원)'].sum()
df1
```

## 사용 목적에서 마지막 단어만 남기기


```python
import re
def splitter(x):
    splitted = re.split(r'\(|\)', x)
    if splitted[-1] == '':
        return splitted[-3]
    else:
        return splitted[-1]

df['사용목적'] = df['사용목적'].apply(lambda x : splitter(x).split()[-2] + ' ' + splitter(x).split()[-1])
df
```

## 사용목적 별 사용 내역 확인

```python
g = df.groupby(['사용목적'])['사용금액(원)'].sum()
g
```

```python
import matplotlib.pyplot as plt
import wordcloud

wc = wordcloud.WordCloud(font_path='C:/Windows/Fonts/malgun.ttf', background_color='black')
cloud = wc.generate_from_frequencies(g)
plt.imshow(cloud)
plt.show()
```
