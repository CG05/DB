# 2. 파일 읽기

```python
import pandas as pd

df_train = pd.read_csv('data/santander-customer-satisfaction/train.csv')
df_test = pd.read_csv('data/santander-customer-satisfaction/test.csv')

df_train.head(2)
```

> var3를 기준으로 훈련 데이터와 검증데이터에 각각 120개 가량의 데이터가 존재한다. 그리고 바로 다음 수치는 0으로 나타난다. 이는 특정 에러가 발생했을 경우 나타났을 것으로 판단하며(보통 개발 시 절대 나타날 수 없는 값을 에러로 사용하는 경우도 있음으로) 가장 빈도가 많은 2로 대체할 예정이다.(4분위수 대다수에 포진됨)

> -999999가 116개 나타난 것을 볼 수 있다. 여기서는 최빈값으로 값을 대체하자

```python
df_train['var3'] = df_train['var3'].replace(-999999, 2)
df_test['var3'] = df_test['var3'].replace(-999999, 2)
```

```python
df_train['var3'].value_counts().sort_index()
```

```python
df_test['var3'].value_counts().sort_index()
```

## 3.3 불필요한 feature 제거

> 첫번째 feature인 ID는 단순 식별자 일 것임으로 예측에 불필요한 데이터가 된다.

# 4. 데이터 분리

```python
y_train = df_train['TARGET']
X_train = df_train.drop('TARGET', axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, random_state=10)
```

# 5. XGBoost

```python
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score, accuracy_score

xgb = XGBClassifier(n_estimators=500, random_state=10)
xgb.set_params(early_stopping_rounds=100, eval_metric='auc')
xgb.fit(X_train, y_train, eval_set=[(X_train, y_train), (X_test, y_test)])
```

```python
xgb.predict_proba(X_test)[:5]
```

```python
y_test
```

```python
pred = xgb.predict(X_test)

accuracy_score(y_test, pred)
```

```python
xgb_roc_score = roc_auc_score(y_test, xgb.predict_proba(X_test)[:, 1])
print('ROC AUC 값 : ', xgb_roc_score)
```

```python
xgb_roc_score = roc_auc_score(y_test, xgb.predict_proba(X_test)[:, 0])
print('ROC AUC 값 : ', xgb_roc_score)
```

- https://developers.google.com/machine-learning/crash-course/classification/roc-and-auc?hl=ko
- https://bskyvision.com/1165

> AUC는 이진데이터 구분 시 사용되며 만족 불만족 두 가지에 대한 평가를 하여 사용하였다.

> predict_proba는 확률을 얻게 되며 보통 1이 될 확률로 값을 구하게 된다.

```python
from sklearn.model_selection import GridSearchCV

xgb = XGBClassifier(n_estimators=100)

params = {
    'max_depth':[3, 6],
    'min_child_weight':[1,3],
    'colsample_bytree':[0.5, 0.75]
}

grid_cv = GridSearchCV(xgb, param_grid=params, cv=3)
grid_cv.fit(X_train, y_train, eval_set=[(X_test, y_test)])

print('최적 파라미터 : ', grid_cv.best_params_)
print('최고 AUC :', roc_auc_score(y_test, grid_cv.predict_proba(X_test)[:, 1]))
```

> 6분 정도 소요

>LightGBM

```python
from lightgbm import LGBMClassifier

lgbm = LGBMClassifier(n_estimators=500)

lgbm.fit(X_train, y_train, eval_set=[(X_test, y_test)])

lgbm_roc_score = roc_auc_score(y_test, lgbm.predict_proba(X_test)[:,1])
print('ROC AUC 값 : ', lgbm_roc_score)
```

튜닝
