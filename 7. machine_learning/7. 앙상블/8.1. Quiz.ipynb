1. 데이터 가져오기

```python
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()
```

2. 학습 및 검증 데이터 분할

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state=10)
```

3. 모델 생성

```python
from xgboost import XGBClassifier

xgb = XGBClassifier(n_estimators=300, learning_rate=0.1, max_depth=3)
# logloss : 이진 분류
# mlogloss : 멀티 클래스 분류
xgb.set_params(early_stopping_rounds=100, eval_metric='mlogloss')
xgb.fit(X_train, y_train,verbose=True, eval_set=[(X_test, y_test)])

pred = xgb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, pred)
```

```python
xgb = XGBClassifier(n_estimators=500, learning_rate=0.1, max_depth=3, min_child_weight=10)
# logloss : 이진 분류
# mlogloss : 멀티 클래스 분류
xgb.set_params(early_stopping_rounds=100, eval_metric='mlogloss')
xgb.fit(X_train, y_train,verbose=True, eval_set=[(X_test, y_test)])

pred = xgb.predict(X_test)
```

```python
accuracy_score(y_test, pred)
```

4. 튜닝

```python
from sklearn.model_selection import GridSearchCV

def best_params(n_estimators_list, max_depth_list, subsample_list):
    param = {
        'n_estimators':n_estimators_list,
        'max_depth':max_depth_list,
        'subsample':subsample_list
    }

    grid_cv = GridSearchCV(xgb, param_grid=param, scoring='accuracy', 
                           verbose=1, n_jobs=-1, cv=3)
    grid_cv.fit(X_train, y_train, eval_set=[(X_test, y_test)])
    print('최고 하이퍼 파라미터 : ', grid_cv.best_params_)
    print('최고 정확도 : ', grid_cv.best_score_)
```

```python
n_estimators_list = range(100, 1001, 100)
max_depth_list = range(1, 52, 10)
subsample_list = [0, 0.3, 0.6, 0.9]

best_params(n_estimators_list, max_depth_list, subsample_list)
```

```python
n_estimators_list = range(50, 201, 50)
max_depth_list = range(1, 10)
subsample_list = [0.7, 0.8, 0.9, 1]

best_params(n_estimators_list, max_depth_list, subsample_list)
```
