#1. 데이터셋 얻기

> 매개변수 `num_words=10000`은 훈련 데이터에서 가장 자주 나타나는 단어 10,000개만 사용하겠다는 의미이다. 드물게 나타나는 단어는 무시하기 위해 사용된다.

> 변수 `train_data`와 `test_data`는 리뷰이며 단어 인덱스의 리스트로 되어 있다. `train_labels`와 `test_labels`는 부정을 나타내는 0과 긍정을 나타내는 1의 리스트이다.

#2. 리뷰 확인

##2.1 리뷰 인덱스 얻기
> imdb에 별도로 저장되어 있는 word index 정보 얻기

##2.2 인덱스 검색을 위환 치환
> 데이터셋이 index 정보로 이루어져 있어 key와 value를 치환한다.

> 0의 None은 데이터가 존재하지 않기 때문에 None으로 표현된 것이다.

##2.3 리뷰 디코딩
> 리뷰의 index 정보는 0, 1, 2는 '패딩', '문서 시작', '사전에 없음'으로 사전 예약 되어있어 초기 값이 3으로 시작해야 한다. 위의 index 정보에서 1이 the로 사용되는데 여기에 3을 더한 4가 실제 index의 정보가 되는 것이다.

> 참고 : https://wikidocs.net/83544

> 결국 데이터 확인을 위해서는 실제 데이터에서 3을 빼고 처리하면 실제이터가 나타나게 된다. 3을 빼는 이유는 앞에서 이야기한 0, 1, 2를 제외하고 처리하기 위함이다. word_index에서 0이하의 수에 대해 None으로 표현되기 때문에 0, 1, 2 모두 None으로 표기 하기 위해 3을 뺀다.

> 첫 번째 리뷰에서 첫 시작은 -2로 나타나며 시작을 의미하게 된다. 중간 내용에는 -1이 나타나게 되는데 이 값이 없는 데이터 이며 실제로 데이터가 없는 것이 아니고 첫 데이터를 load할 때 num_words=10000을 처리하여 고유 데이터를 제거했기 때문에 없는 단어로 표기된 것이다. num_words를 100000으로 변경한 후 확이해 보면 rebert 다음 단어가 redford's인 것을 확인할 수 있다.
- 667 robert
- 22662 redford's
- 6 is


> 결국 변경된 값을 '?'로 표기하여 위와 같은 결과를 얻을 수 있다.

#3. 벡터 변환
> 신경망에서는 리스트를 텐서로 변환하여 사용해야 하며 다음과 같이 두 가지 방법으로 변환할 수 있다.
- 같은 길이가 되도록 리스트에 패딩을 추가하고 (samples, sequence_length) 크기의 정수 텐서로 변환한다. 그 다음 이 정수 텐서를 다룰 수 있는 층을 신경망의 첫 번째 층으로 사용한다(Embedding 층을 말하며 나중에 확이해 보자)
- 리스트를 원-핫 인코딩하여 0과 1의 벡터로 변환한다. 그 다음 부동 소수 벡터 데이터를 다룰 수 있는 Dense 층을 신경망의 첫 번째 층으로 사용한다.

> 여기서는 두 번째 방법을 사용할 것이다.
