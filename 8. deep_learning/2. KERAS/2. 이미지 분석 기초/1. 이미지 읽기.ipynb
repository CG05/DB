# 2. 동영상 불러오기

## 2.1 동영상 정보 확인

````python
movie_path = 'mov/mov01.avi'
mov = cv2.VideoCapture(movie_path)

width = mov.get(cv2.CAP_PROP_FRAME_WIDTH)
height = mov.get(cv2.CAP_PROP_FRAME_HEIGHT)
count = mov.get(cv2.CAP_PROP_FRAME_COUNT)
fps = mov.get(cv2.CAP_PROP_FPS)
````

````python
print('가로 :', width)
print('세로 :', height)
print('프레임 수 :', count)
print('초당 프레임 수 :', fps)
````

## 2.2 동영상 동작시키기

````python
if not mov.isOpened():
    print('Error : Could not open video file')
    exit()

for i in range(10):
    ret, frame = mov.read()
    
    if not ret :
        print('Error : Could not read frame')
        break
    
    cv2.imshow('Frame', frame)
    if cv2.waitKey(1000) & 0xFF == ord('q'):
        break

mov.release()
cv2.destroyAllWindows()
````

> GPU를 이용했지만 큰 차이는 없으며 생각보다 많이 느리다.

# 3. 동영상을 이미지로 저장하기

````python
import cv2 

movie_path = 'mov/mov01.avi'
mov = cv2.VideoCapture(movie_path)

cnt = 0 

for i in range(10):
    mov.isOpened()
    ret, frame = mov.read()

    if ret :
        print(cnt)
        file_path = 'snapshot/snapshot_'+str(cnt)+'.jpg'
        cv2.imwrite(file_path, frame)
        cnt+=1

mov.release()
cv2.destroyAllWindows()
````

````python
import cv2 

movie_path = 'mov/mov02.avi'
mov = cv2.VideoCapture(movie_path)

cnt = 0 

for i in range(10):
    mov.isOpened()
    ret, frame = mov.read()

    if ret :
        print(cnt)
        file_path = 'download_capture/snapshot_02_'+str(cnt)+'.jpg'
        cv2.imwrite(file_path, frame)
        cnt+=1

mov.release()
cv2.destroyAllWindows()
````
