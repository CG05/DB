```python
from keras.models import load_model

model = load_model('model/mnist240717.keras')
```

```python
import cv2

img = cv2.imread('model/3.png')
```

```python
import matplotlib.pyplot as plt

plt.imshow(img, cmap=plt.cm.binary)
plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap=plt.cm.binary)
plt.show()
```

```python
gray.shape
```

```python
gray_reshape = gray.reshape(1, 28*28)
gray_reshape = gray_reshape.astype('float32') / 255
```

```python
pred = model.predict(gray_reshape)
pred[0].argmin() + 1
```
