# 1. 모델 만들기
> 모델을 최초 2개만 진행하고 하나씩 추가 하면서 결과를 확인해 보자.

> Conv2D는 다음과 같은 파라미터를 갖는다.
- filters : 32의 출력 결과를 얻게 된다.
- kernel_size : (3, 3)의 kernel을 사용하여 filtering하게 된다.
- activation : relu라는 활성화 함수를 이용하게 된다.
- input_shape : (28, 28, 1)의 입력 데이터를 이용한다.

image.png

> 28개의 입력 데이터가 3,3의 kernel을 들려 결과를 얻어 내면 데이터가 중복되게 처리 되어 다음과 같은 결과를 얻는다.
- 28(input_size) - 3(kernel-size) = 26(output_size)

> 위의 그림을 예로 들면 8, 8 데이터에서 3,3의 kernel로 필터를 하게 되면 6, 6의 결과가 나타나게 된다.

image.png

> MaxPooling2D는 지정간 크기만큼의 공간에서 가장 큰 값을 추출하게 된다. 위의 그림에서 6,6의 공간을 2,2 공간으로 분리하여 최대값을 추출하면 3, 3 공간으로 변환하게 된다. 즉, 원래 크기의 반으로 줄게 되는 것이다.

> 지금까지 내용을 간단하게 summary()를 통해 확인할 수 있다.

> 3, 3, 64형식의 데이터를 Flatten을 이용하여 1차원으로 변경한다.
