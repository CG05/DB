### 문제 4 
X = [ [1,2,3], [4,5,6]] , y = [ [7,8,9],[10,11,12] ] x와 y를 연결해서 [ [1,2,3,7,8,9], [4,5,6,10,11,12] ] 를 만드세요



```python
x = np.array([[1,2,3],[4,5,6]])
y = np.array([[7,8,9],[10,11,12]])
np.concatenate((x,y), axis=1)
```

### 문제 6
x 가 [0,1,2] 일때, [0,0,1,1,2,2] 를 생성하세요
