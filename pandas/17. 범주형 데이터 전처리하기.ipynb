# One-hot encoding
범주형(Categorical) 데이터 전처리하기

- 범주형 데이터는 연산이 불가능 : 문자열일 수 있기 때문에 ...
- one-hot encoding : 범주형 데이터의 값을 컬럼으로 변경하여 데이터를 숫자로 표현하는 방식

```
color = ['red', 'blue', 'black', 'red', 'black']
```
--------------
One-hot encoding을 처리하면
```
red = [1,0,0,1,0]
blue = [0,1,0,0,0]
black = [0,0,1,0,1]
```

## One-hot encoding
 - 범주형 데이터는 분석단계에서 계산이 어렵기 때문에 숫자형으로 변경이 필요함
 - 범주형 데이터의 각 범주(category)를 column레벨로 변경
 - 해당 범주에 해당하면 1, 아니면 0으로 채우는 인코딩 기법
 - pandas.get_dummies 함수 사용
   - drop_first : 첫번째 카테고리 값은 사용하지 않음

````python
# Pandas.get_dummies() : 문자형 데이터를 one-hot encoding으로 변환
# Name, Cabin, Embarked, Ticket을 처리해버림
# 문자형 데이터를 가지고 있는 열을 모두 one-hot encoding으로 변환
# 문제는 문자형 데이터가 너무 많으면 열이 너무 많아짐
pd.get_dummies(df)
````

````python
pd.get_dummies(df, columns=['Pclass', 'Sex', 'Embarked'])
````

````python
# Pandas.get_dummies(데이터, columns=[변환할 열들], drop_first=(첫 컬럼 필요한 지 아닌지 True/False))

pd.get_dummies(df, columns=['Pclass', 'Sex', 'Embarked'], drop_first=True)
# Pclass_1, Sex_female, Embarked_C 컬럼이 사라짐 : male이 아니면 female이므로 둘 중 한 열은 필요가 없음
````

````python
# 머신러닝을 하다 보면 범주형 데이터를 수치화했을 때 1, 0으로 나올 때가 대부분
# 하지만 가끔 0.98, 0.01 이런 식으로 나오는 경우도 있음
# 그럴 때는 1에 가까운것이 1, 0에 가까운 것이 0으로 바꾸는 것이 좋음
````
