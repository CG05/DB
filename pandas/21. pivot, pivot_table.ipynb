# pivot, pd.pivot_table 함수

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    '지역': ['서울', '서울', '서울', '경기', '경기', '부산', '서울', '서울', '부산', '경기', '경기', '경기'],
    '요일': ['월요일', '화요일', '수요일', '월요일', '화요일', '월요일', '목요일', '금요일', '화요일', '수요일', '목요일', '금요일'],
    '강수량': [100, 80, 1000, 200, 200, 100, 50, 100, 200, 100, 50, 100],
    '강수확률': [80, 70, 90, 10, 20, 30, 50, 90, 20, 80, 50, 10]
                  })

df
```

## transpose(),  T
- row ↔ column 이 바뀐 형태 
- 전치행렬

```python
df.T
```

# pivot() 
 - dataframe의 형태를 변경
 - 인덱스, 컬럼, 데이터로 사용할 컬럼을 명시

```python
df.pivot(index='지역', columns='요일')
```

```python
df.loc[len(df)] = pd.Series(['경기', '금요일', 111, 11], index = ['지역', '요일', '강수량', '강수확률'])
df
```

```python
df.pivot(index='지역', columns='요일')
```

```python
df.drop(len(df)-1, axis=0, inplace=True)
df
```

```python
df.pivot(index='요일', columns='지역')
```

```python
df.pivot(index='요일', columns='지역', values='강수량')
```

# pd.pivot_table()
 - 기능적으로 pivot과 동일
 - pivot과의 차이점
   - 중복되는 모호한 값이 있을 경우, aggregation 함수 사용하여 값을 채움

```python
pd.pivot_table(df, index='지역', columns='요일')
```

```python
df2 = pd.DataFrame({
    '지역': ['서울', '서울', '서울', '경기', '경기', '부산', '서울', '서울', '부산', '경기', '경기', '경기'],
    '요일': ['월요일', '월요일', '수요일', '월요일', '화요일', '월요일', '목요일', '금요일', '화요일', '수요일', '목요일', '금요일'],
    '강수량': [100, 80, 1000, 200, 200, 100, 50, 100, 200, 100, 50, 100],
    '강수확률': [80, 70, 90, 10, 20, 30, 50, 90, 20, 80, 50, 10]
                  })

df2
```

```python
df2.pivot(index='지역', columns='요일')
```

```python
pd.pivot_table(df2, index='지역', columns='요일')
# 중복값의 평균값으로 대체
```

```python
pd.pivot_table(df2, index='지역', columns='요일', aggfunc=np.mean)
```

```python
def zero(data):
    print("data : ", data)
    data = pd.Series(map(lambda x: 0 if np.isnan(x) else x, data))
    return data
```

```python
pd.pivot_table(df2, index='지역', columns='요일', aggfunc=zero).fillna(0)
```
